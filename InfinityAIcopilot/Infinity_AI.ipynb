# Infinity AI Copilot Overview

Infinity AI Copilot is Check Point's new Generative AI Security Assistant, designed to enhance cybersecurity operations across various Check Point products. It serves as an integrated tool within the Infinity portal, facilitating a wide range of security tasks from log analysis to policy adjustments. 

> **Note:** Access to Infinity AI Copilot is currently limited; interested parties should refer to the [Check Point website](https://pages.checkpoint.com/2024-feb-infinity-ai-copilot-preview-xdr-xpr.html) or contact support for availability.

## Product Integration

Infinity AI Copilot seamlessly integrates with the following Check Point products:

### Smart-1 Cloud
- Manages Check Point gateways within the Infinity portal environment.

### Harmony Endpoint
- Protects endpoints across various OS, enhancing security and operational efficiency.

## Features and Capabilities

Upon accessing the Infinity portal, users can interact with the Infinity AI Copilot via a dedicated button. The AI Copilot can assist with various queries, from troubleshooting VPN issues to identifying vulnerabilities like Log4j and assessing network attack incidents. It also provides insights into user activities and policy enforcement, such as tracking TikTok usage or explaining access denials.

## Policy Management

For policy adjustments, such as modifying access rules, Infinity AI Copilot requires write access permissions. This ensures secure and controlled changes to security policies.
